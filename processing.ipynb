```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load the four datasets which will be merged into one dataset for analysis

df_species_class = pd.read_csv("scmpaintertidalclasstable20150220.csv")
df_site_table = pd.read_csv("scmpaintertidalsitetable20150220.csv")
df_counts = pd.read_csv("sc_mpa_marine_species_counts_2015_0505.csv")
df_sizes = pd.read_csv("sc_mpa_marine_species_sizes_2015_0220.csv")
```

```python
for col in df_counts.columns:
    print(f"{col}: {df_counts[col].apply(type).nunique()} unique types")
```

```python
df_counts['PlotID'] = df_counts['PlotID'].astype(str)
```

```python
df_species_class.columns
```

```python
df_species_class = df_species_class.drop(columns=[
    'SC_project_short_code', 
    'preliminary_classification_OriginalName', 
    'class_code',
    'kingdom', 
    'phylum_division', 
    'class', 
    'order',
    'family', 
    'genus', 
    'species'
])
```

```python
df_species_class = df_species_class.dropna(subset=['SixLetterCode'])
df_species_class = df_species_class.rename(columns={"final_classification_NameConvergence": "SpeciesName"})
```

```python
df_species_class
```

```python
df_counts = df_counts.merge(df_species_class, on="SixLetterCode", how="left")
df_counts.head()
```

```python
for col in df_counts.columns:
    print(f"{col}: {df_counts[col].nunique()} unique values")
```

```python
df_counts = df_counts.drop(columns = [
    "SC_project_short_code",
    "SampleSeason",
    "SixLetterCode"
])
df_counts.head()
```

```python
df_sizes = df_sizes.merge(df_species_class, on="SixLetterCode", how="left")
df_sizes.head()
```

```python
for col in df_sizes.columns:
    print(f"{col}: {df_sizes[col].nunique()} unique values")
```
